```
# Heston Model Option Pricing â€“ Simulation Analysis

import numpy as np
import matplotlib.pyplot as plt
from src.heston_model import heston_option_price, simulate_heston_paths

# Parameters
S0 = 67.0       # initial asset price
K = 70.0        # strike price
T = 0.5         # time to maturity (in years)
r = 0.05        # risk-free rate
v0 = 0.04       # initial variance
kappa = 1.5     # mean reversion rate
theta = 0.04    # long-run variance
sigma = 0.3     # vol of vol
rho = -0.7      # correlation
N = 252         # time steps
M = 10000       # simulations

# Run pricing
call_price = heston_option_price(S0, K, T, r, v0, kappa, theta, sigma, rho, option_type="call", N=N, M=M)
print(f"Heston Model Call Option Price: {call_price:.4f}")

# Simulate paths
S_paths, v_paths = simulate_heston_paths(S0, v0, r, kappa, theta, sigma, rho, T, N, M=100)

# Plot asset paths
plt.figure(figsize=(10, 5))
for i in range(100):
    plt.plot(np.linspace(0, T, N + 1), S_paths[i])
plt.title("Heston Simulated Asset Price Paths-USO")
plt.xlabel("Time (years)")
plt.ylabel("Asset Price")
plt.grid(True)
plt.tight_layout()
plt.savefig(r"C:\Users\adity\OneDrive\Documents\GitHub\Stochastic-Volatility-Option-Pricing-using-the-Heston-Model\plots/heston_asset_paths_USO.png")
plt.show()

# Plot volatility paths
plt.figure(figsize=(10, 5))
for i in range(100):
    plt.plot(np.linspace(0, T, N + 1), v_paths[i])
plt.title("Heston Simulated Volatility Paths-USO")
plt.xlabel("Time (years)")
plt.ylabel("Variance")
plt.grid(True)
plt.tight_layout()
plt.savefig(r"C:\Users\adity\OneDrive\Documents\GitHub\Stochastic-Volatility-Option-Pricing-using-the-Heston-Model\plots/heston_Volatility_paths_USO.png")
plt.show()
```
